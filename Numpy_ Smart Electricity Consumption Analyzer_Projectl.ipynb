## Smart Electricity Consumption Analyzer (Using NumPy Arrays)

Used
- Python
- NumPy (arrays only)

Data Assumption
- 1 month = 30 days
- Daily electricity units consumed

```python
units = np.array([
    5.2, 6.1, 5.8, 7.0, 8.5,
    9.2, 6.4, 5.9, 7.8, 8.1,
    6.5, 7.3, 9.0, 10.2, 8.9,
    7.6, 6.8, 5.5, 6.0, 7.1,
    8.4, 9.6, 10.5, 9.8, 8.2,
    7.0, 6.3, 5.9, 6.7, 7.4
])                               # 1D Array

print("Daily Units:", units)
```

#### Total & Average Consumption

```python
total_units = np.sum(units) # sum of all units 
avg_units = np.mean(units)   # average units
med_units = np.median(units)   # median units
min_units = np.min(units)   # minimum units
max_units = np.max(units)   # maximum units

print("Total Units in Month:", total_units)
print("Average of Daily Units:", avg_units)
print("Median of Daily Units:", med_units)
print("Minimum of Daily Units:", min_units)
print("Maximum of Daily Units:", max_units)

```

- Minimum unit is 5.2 , and maximum is 10.5.
- the average unit is around 7.48.
- the month units is 224.7.


#### High Usage Days (>9 units)

```python
high_usage_days = np.where(units > 9)

print("High Usage Days:", high_usage_days[0] + 1)
```

#### Wastage Detection
- (If usage > 20% above average)

```python
wastage_days = np.where(units > avg_units * 1.2)

print("Possible Wastage Days:", wastage_days[0] + 1)
```

#### Electricity Bill Estimation

- Assume:
1 unit = ₹6

```python
bill_amount = total_units * 6
print("Estimated Electricity Bill: ₹", bill_amount)
```

#### Best & Worst Day

```python
best_day = np.argmin(units) + 1
worst_day = np.argmax(units) + 1

print("Lowest Consumption Day:", best_day)
print("Highest Consumption Day:", worst_day)
```
